```python
import scipy.io as sio
import numpy as np

# 加载 .mat 文件
data_mat1 = sio.loadmat('data/adam.mat')
data_mat2 = sio.loadmat('data/new_Quake_10x_Spleen_sort.mat') 
```

```python
# 查看文件中包含的所有键
print("adam可用的键:", data_mat1.keys())
print("my data可用的键:", data_mat2.keys())
```

```python


# 假设 data_mat2['fea'] 是一个 NumPy 数组或可以转换为 NumPy 数组的对象
data_matrix = data_mat2['fea']

# 计算每行非零数据的数量
non_zero_counts = np.count_nonzero(data_matrix, axis=1)

# 计算每行大于1的数据个数
greater_than_one_counts = np.sum(data_matrix > 1, axis=1)

# 计算每行的最大值、最小值和平均值
non_zero_max = np.max(data_matrix, axis=1)
non_zero_min = np.min(data_matrix, axis=1)
non_zero_mean = np.where(non_zero_counts > 0, np.sum(data_matrix, axis=1) / non_zero_counts, 0)

# 输出结果

print("每行非零数据数量:", np.mean(non_zero_counts))

print("每行大于1数据个数:", np.mean(greater_than_one_counts))
print("每行最大值:", np.mean(non_zero_max))
print("每行最小值:", np.mean(non_zero_min))
print("每行平均值:", np.mean(non_zero_mean))

```
